# Importing Libraries

```python
import pandas as pd
from pycaret.classification import *
import pickle
```

```python
# Loading dataset
data = pd.read_csv('train.csv')
```

```python
data
```

```python
# Setting up the PyCaret environment
clf_setup = setup(data, target='Survived')
```

```python
# Compare different models
best_model = compare_models()
```

```python
lr = create_model('lr')
```

```python
# Tune hyperparameters of the model
tuned_dt= tune_model(lr)
```

```python
data1=pd.read_csv('test.csv')
```

```python
data1
```

```python
# Make predictions
predictions = predict_model(tuned_dt, data=data1)
```

```python
predictions
```

```python
# Finalize the model
final_dt = finalize_model(tuned_dt)

# Save the model
save_model(final_dt, 'saved_model')
```
